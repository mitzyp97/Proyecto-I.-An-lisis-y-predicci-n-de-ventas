#Exploración de Datos:
Calcula el total de ventas por categoría de producto.

Calcula el promedio de ventas diarias por categoría de producto.

Identifica las categorías de productos con mayores y menores ventas.

```python
ventas_categoria = df.groupby("Product Category")["Total Amount"].sum()

print("Total de ventas por categoría:")
ventas_categoria
```

```python
promedio_diario = df.groupby(["Date", "Product Category"])["Total Amount"].sum().unstack().mean().round(2)
#unstack() para convertir "Product Category" en columnas.

###promedio_diario = df.groupby(["Date", "Product Category"])["Total Amount"].sum().mean().round(2)

print("Promedio de ventas diarias por categoría de producto:")
promedio_diario
```

```python
categoria_mayor_venta = ventas_categoria.idxmax()    # idxmax() devuelve la categoría con el valor más alto.
monto_mayor_venta = ventas_categoria.max()           # max() devuelve el valor más alto.

print(f"Categoría con mayores ventas: {categoria_mayor_venta} {monto_mayor_venta}")
```

```python
categoria_menor_venta = ventas_categoria.idxmin()    # idxmin() devuelve la categoría con el valor más bajo.
monto_menor_venta = ventas_categoria.min()           # min() devuelve el valor más bajo.

print(f"Categoría con menores ventas: {categoria_menor_venta} {monto_menor_venta}")
```

#Manipulación de Datos:
Filtra los datos para mostrar solo las ventas de una categoría de producto específica.

Realiza operaciones de suma, resta, multiplicación y división en los datos para obtener estadísticas adicionales.

```python
#Eligiendo la categoria Beauty

categoria = "Beauty"
ventas_filtradas = df[df["Product Category"] == categoria]

print(f"Ventas filtradas para la categoría: {categoria}")
ventas_filtradas
```

```python
# Suma total de productos vendidos

total_productos = df["Quantity"].sum()
total_productos
```

```python
# calcular cuánto inventario queda luego de cada venta, suponiendo que inicialmente cada producto tiene un stock de 100 unidades

df["Inventario Restante"] = 100 - df["Quantity"]

df[["Quantity", "Inventario Restante"]].head()
```

```python
# Descuento simulado del 10% sobre cada venta

df["Monto con Descuento 10%"] = df["Total Amount"] * 0.9

df[["Total Amount", "Monto con Descuento 10%"]].head()
```

```python
# Precio promedio por unidad vendida

df["Precio Promedio por Unidad"] = df["Total Amount"] / df["Quantity"]

df[["Total Amount", "Quantity", "Precio Promedio por Unidad"]].head()
```
